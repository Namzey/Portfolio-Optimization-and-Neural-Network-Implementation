```python
import yfinance as yf
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import statsmodels.api as sm

torch.manual_seed(42)
np.random.seed(42)


# 1. DOWNLOAD DATA & EMPIRICALLY CALIBRATE HESTON PARAMETERS

print("Downloading data for calibration...")
data = yf.download(['SPY', '^VIX'], start='2000-01-01', end='2024-01-01')['Close']
df = pd.DataFrame({
    'S': data['SPY'],
    'v': (data['^VIX'] / 100) ** 2  # Variance is squared VIX
}).dropna()

dt = 1 / 252

df['S_ret'] = df['S'].pct_change()
df['v_diff'] = df['v'].diff()
df = df.dropna()

#Estimate Parameters
mu_est = df['S_ret'].mean() / dt
rho_est = df['S_ret'].corr(df['v_diff'])
theta_est = df['v'].mean()

# Align data for OLS to find kappa and sigma_v
reg_df = pd.DataFrame({
    'y': df['v_diff'],
    'x': df['v'].shift(1)
}).dropna()

y_vals = reg_df['y'].values
X_vals = sm.add_constant(reg_df['x'].values)

ols_model = sm.OLS(y_vals, X_vals).fit()
alpha, beta = ols_model.params

kappa_est = -beta / dt
residuals = ols_model.resid
v_prev = reg_df['x'].values
sigma_v_est = np.sqrt(np.var(residuals / np.sqrt(v_prev + 1e-8)) / dt)

print("\n--- EMPIRICALLY CALIBRATED HESTON PARAMETERS ---")
print(f"Drift (mu):                  {mu_est:.4f}")
print(f"Mean Reversion (kappa):      {kappa_est:.4f}")
print(f"Long-Term Variance (theta):  {theta_est:.4f}")
print(f"Vol of Vol (sigma_v):        {sigma_v_est:.4f}")
print(f"Correlation (rho):           {rho_est:.4f}")
print("------------------------------------------------\n")


# 2. CONTINUOUS-TIME SDE SIMULATION (EMPIRICALLY CALIBRATED)

def simulate_heston(
    S0=1.0,
    v0=theta_est,
    mu=mu_est,
    kappa=kappa_est,
    theta=theta_est,
    sigma_v=sigma_v_est,
    rho=rho_est,
    r=0.04,
    T=1.0, steps=252, paths=3000
):
    dt_sim = T / steps
    S = np.zeros((steps+1, paths))
    v = np.zeros((steps+1, paths))
    B = np.zeros((steps+1, paths))

    S[0], v[0], B[0] = S0, v0, 1.0

    Z1 = np.random.normal(size=(steps, paths))
    Z2 = np.random.normal(size=(steps, paths))

    for t in range(steps):
        v_plus = np.maximum(v[t], 0)
        vol = np.sqrt(v_plus)

        dW_S = np.sqrt(dt_sim) * Z1[t]
        dW_v = np.sqrt(dt_sim) * (rho*Z1[t] + np.sqrt(1-rho**2)*Z2[t])

        S[t+1] = S[t] * (1 + mu*dt_sim + vol*dW_S)
        v[t+1] = v[t] + kappa*(theta - v_plus)*dt_sim + sigma_v*vol*dW_v
        B[t+1] = B[t] * (1 + r*dt_sim)

    return S, v, B

print("Simulating Empirically Calibrated Heston SDEs...")
S, v, B = simulate_heston()


# 3. ANN HJB SOLVERS (Soft Regime vs. No Regime)


class PolicyANN_Regime(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 16), nn.SiLU(),
            nn.Linear(16, 16), nn.SiLU(),
            nn.Linear(16, 1), nn.Sigmoid()
        )
    def forward(self, t, var):
        x = torch.stack([t, var], dim=1)
        return self.net(x).squeeze()

class PolicyANN_NoRegime(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 16), nn.SiLU(),
            nn.Linear(16, 16), nn.SiLU(),
            nn.Linear(16, 1), nn.Sigmoid()
        )
    def forward(self, t):
        return self.net(t.unsqueeze(1)).squeeze()

model_regime = PolicyANN_Regime()
model_noregime = PolicyANN_NoRegime()

opt_regime = optim.Adam(model_regime.parameters(), lr=0.005)
opt_noregime = optim.Adam(model_noregime.parameters(), lr=0.005)

S_t = torch.tensor(S, dtype=torch.float32)
v_t = torch.tensor(v, dtype=torch.float32)
B_t = torch.tensor(B, dtype=torch.float32)

v_mean = v_t.mean().item()
v_std = v_t.std().item() + 1e-8
fee_rate = 0.001

print("Training ANNs with Path-Dependent Transaction Costs...")
for epoch in range(150):
    opt_regime.zero_grad()
    opt_noregime.zero_grad()

    W_reg = torch.ones(S.shape[1])
    W_noreg = torch.ones(S.shape[1])

    pi_prev_reg = torch.full((S.shape[1],), 0.5)
    pi_prev_noreg = torch.full((S.shape[1],), 0.5)

    for t in range(252):
        t_input = torch.full((S.shape[1],), t*dt)
        var_input = (v_t[t] - v_mean) / v_std

        pi_reg = model_regime(t_input, var_input)
        pi_noreg = model_noregime(t_input)

        cost_reg = torch.abs(pi_reg - pi_prev_reg) * fee_rate
        cost_noreg = torch.abs(pi_noreg - pi_prev_noreg) * fee_rate

        stock_ret = (S_t[t+1] - S_t[t]) / S_t[t]
        bond_ret = (B_t[t+1] - B_t[t]) / B_t[t]

        W_reg = W_reg * (1 + pi_reg*stock_ret + (1-pi_reg)*bond_ret - cost_reg)
        W_reg = torch.clamp(W_reg, min=1e-6)

        W_noreg = W_noreg * (1 + pi_noreg*stock_ret + (1-pi_noreg)*bond_ret - cost_noreg)
        W_noreg = torch.clamp(W_noreg, min=1e-6)

        pi_prev_reg = pi_reg
        pi_prev_noreg = pi_noreg

    loss_reg = -torch.mean(torch.log(W_reg))
    loss_noreg = -torch.mean(torch.log(W_noreg))
    loss_reg.backward()
    loss_noreg.backward()
    opt_regime.step()
    opt_noregime.step()


# 4. REALISTIC TRADING ENGINE

def run_realistic_trading(target_weights, stock_ret, bond_ret, fee_rate=0.001, tax_rate=0.15, threshold=0.05):
    n = len(stock_ret)
    stock_val = 0.5
    bond_val = 0.5
    cost_basis = 0.5

    returns = []
    actual_weights = []
    trades_executed = 0
    prev_total_tracker = 1.0

    for i in range(n):
        stock_val *= (1 + stock_ret.iloc[i])
        bond_val *= (1 + bond_ret.iloc[i])

        total_val = stock_val + bond_val
        current_weight = stock_val / total_val
        pi = target_weights[i]

        if abs(pi - current_weight) > threshold:
            trades_executed += 1
            target_stock = total_val * pi

            if pi > current_weight:
                buy_amt = target_stock - stock_val
                fee = buy_amt * fee_rate
                bond_val -= (buy_amt + fee)
                stock_val += buy_amt
                cost_basis += buy_amt

            elif pi < current_weight:
                sell_amt = stock_val - target_stock
                fee = sell_amt * fee_rate
                proportion = sell_amt / stock_val if stock_val > 0 else 0
                cost_portion = cost_basis * proportion
                profit = sell_amt - cost_portion
                tax = profit * tax_rate if profit > 0 else 0

                bond_val += (sell_amt - fee - tax)
                stock_val -= sell_amt
                cost_basis -= cost_portion

        new_total = stock_val + bond_val
        port_ret = (new_total - prev_total_tracker) / prev_total_tracker
        returns.append(port_ret)
        actual_weights.append(stock_val / new_total)

        prev_total_tracker = new_total

    return np.array(returns), np.array(actual_weights), trades_executed

```

```python

# 5. REAL DATA BACKTEST & TARGET WEIGHT GENERATION

print("Downloading Real Market Data for Backtest...")
data_bt = yf.download(['SPY', 'TLT', '^VIX'], start='2007-01-01', end='2024-01-01')['Close']
df_bt = pd.DataFrame({
    'spy_ret': data_bt['SPY'].pct_change(),
    'tlt_ret': data_bt['TLT'].pct_change(),
    'vix': data_bt['^VIX']
}).dropna()

# Smooth the VIX to prevent churn (Overfitting to noise)
smoothed_vix = df_bt['vix'].rolling(window=10).mean().bfill()
variance_real = (smoothed_vix / 100) ** 2
variance_norm = (variance_real - v_mean) / v_std

target_weights_regime = []
target_weights_noregime = []

model_regime.eval()
model_noregime.eval()

with torch.no_grad():
    for i in range(len(df_bt)):
        # SAWTOOTH FIX: Lock time horizon to 0.0 for infinite-horizon behavior
        t_locked = 0.0

        target_weights_regime.append(model_regime(
            torch.tensor([t_locked], dtype=torch.float32),
            torch.tensor([variance_norm.iloc[i]], dtype=torch.float32)
        ).item())

        target_weights_noregime.append(model_noregime(
            torch.tensor([t_locked], dtype=torch.float32)
        ).item())

# Run both target paths through the realistic trading simulator

ret_regime, act_w_regime, trades_regime = run_realistic_trading(target_weights_regime, df_bt['spy_ret'], df_bt['tlt_ret'])
ret_noregime, act_w_noregime, trades_noregime = run_realistic_trading(target_weights_noregime, df_bt['spy_ret'], df_bt['tlt_ret'])
```

```python

# 6. REPORTING SUCCESS RATIOS & PLOTS


def calculate_precise_metrics(returns):
    cum_returns = (1 + returns).cumprod()
    annualized_return = cum_returns[-1] ** (252 / len(returns)) - 1
    annualized_vol = np.std(returns) * np.sqrt(252)
    sharpe_ratio = annualized_return / annualized_vol

    rolling_max = np.maximum.accumulate(cum_returns)
    drawdowns = (cum_returns - rolling_max) / rolling_max
    max_drawdown = np.min(drawdowns)

    return annualized_return, annualized_vol, sharpe_ratio, max_drawdown, cum_returns, drawdowns

# Calculate metrics for all 3 strategies
ann_ret_r, vol_r, sharpe_r, dd_r, cum_r, draw_r = calculate_precise_metrics(ret_regime)
ann_ret_nr, vol_nr, sharpe_nr, dd_nr, cum_nr, draw_nr = calculate_precise_metrics(ret_noregime)
ann_ret_q, vol_q, sharpe_q, dd_q, cum_q, draw_q = calculate_precise_metrics(df_bt['spy_ret'].values)

print("\n" + "="*80)
print("EXPLICIT STRATEGY COMPARISON (OUT-OF-SAMPLE BACKTEST)")
print("="*80)
print(f"{'Metric':<25} | {'Soft-Regime ANN':<15} | {'No-Regime ANN':<15} | {'SPY Benchmark':<15}")
print("-" * 80)
print(f"{'Trades Executed':<25} | {trades_regime:<15} | {trades_noregime:<15} | {'0 (Frictionless)':<15}")
print(f"{'Annualized Return':<25} | {ann_ret_r:>15.6f} | {ann_ret_nr:>15.6f} | {ann_ret_q:>15.6f}")
print(f"{'Annualized Volatility':<25} | {vol_r:>15.6f} | {vol_nr:>15.6f} | {vol_q:>15.6f}")
print(f"{'Max Drawdown':<25} | {dd_r:>15.6f} | {dd_nr:>15.6f} | {dd_q:>15.6f}")
print(f"{'Sharpe Ratio':<25} | {sharpe_r:>15.6f} | {sharpe_nr:>15.6f} | {sharpe_q:>15.6f}")
print("="*80)

# 3-Panel Plot for the Report
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 10), gridspec_kw={'height_ratios': [2, 1.5, 1.5]}, sharex=True)

# Top Panel: Cumulative Returns (Log Scale)
ax1.plot(df_bt.index, cum_r, label=f"Soft Regime ANN (Sharpe: {sharpe_r:.2f})", color='blue', linewidth=2)
ax1.plot(df_bt.index, cum_nr, label=f"No-Regime ANN (Sharpe: {sharpe_nr:.2f})", color='orange', linestyle='--', linewidth=2)
ax1.plot(df_bt.index, cum_q, label=f"SPY Benchmark (Sharpe: {sharpe_q:.2f})", color='green', alpha=0.6, linewidth=1.5)
ax1.set_title("Wealth Accumulation: ANN Strategies vs. Constant Market Policy", fontsize=12)
ax1.set_yscale('log')
ax1.set_ylabel("Wealth (Log Scale)")
ax1.legend(loc='upper left')
ax1.grid(alpha=0.3)

# Middle Panel: Drawdowns (Capital Preservation)
ax2.plot(df_bt.index, draw_r, label=f"Soft Regime ANN (Max DD: {dd_r:.1%})", color='blue', linewidth=1.5)
ax2.plot(df_bt.index, draw_nr, label=f"No-Regime ANN (Max DD: {dd_nr:.1%})", color='orange', linestyle='--', linewidth=1.5)
ax2.plot(df_bt.index, draw_q, label=f"SPY Benchmark (Max DD: {dd_q:.1%})", color='green', alpha=0.5, linewidth=1)
ax2.fill_between(df_bt.index, draw_r, 0, color='blue', alpha=0.1)
ax2.fill_between(df_bt.index, draw_nr, 0, color='orange', alpha=0.1)
ax2.fill_between(df_bt.index, draw_q, 0, color='green', alpha=0.1)
ax2.set_title("Drawdown Profile (Capital Preservation)", fontsize=12)
ax2.set_ylabel("Drawdown %")
ax2.legend(loc='lower right')
ax2.grid(alpha=0.3)

# Bottom Panel: Dynamic ANN Target Weight vs VIX
ax3.plot(df_bt.index, target_weights_regime, label="Soft Regime Target ($\pi$)", color='purple', linewidth=1.5)
ax3.plot(df_bt.index, target_weights_noregime, label="No-Regime Target ($\pi$)", color='orange', linestyle='--', alpha=0.7)
ax3.set_ylabel("Target SPY Weight ($\pi$)")
ax3.set_ylim(0, 1.05)
ax3.legend(loc='upper left')

ax4 = ax3.twinx()
ax4.plot(df_bt.index, df_bt['vix'], color='red', alpha=0.3, label='VIX Index')
ax4.set_ylabel("VIX", color='red')
ax4.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# ==============================================================================
# BONUS BLOCK 1: THE UNCONSTRAINED "HEDGE FUND" MODEL (ALLOWS LEVERAGE)
# ==============================================================================
class PolicyANN_Leveraged(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(2, 16), nn.SiLU(),
            nn.Linear(16, 16), nn.SiLU(),
            nn.Linear(16, 1)
        )
    def forward(self, t, var):
        x = torch.stack([t, var], dim=1)
        # Multiply Sigmoid by 3 to allow up to 300% equity (2x margin borrowing)
        return torch.sigmoid(self.net(x)).squeeze() * 3.0

model_lev = PolicyANN_Leveraged()
opt_lev = optim.Adam(model_lev.parameters(), lr=0.005)

print("Training Unconstrained Model (Allowing up to 3x Leverage)...")
for epoch in range(150):
    opt_lev.zero_grad()
    W_lev = torch.ones(S.shape[1])
    pi_prev_lev = torch.full((S.shape[1],), 0.5)

    for t in range(252):
        t_input = torch.full((S.shape[1],), t*dt)
        var_input = (v_t[t] - v_mean) / v_std

        pi_lev = model_lev(t_input, var_input)
        cost_lev = torch.abs(pi_lev - pi_prev_lev) * fee_rate

        stock_ret = (S_t[t+1] - S_t[t]) / S_t[t]
        bond_ret = (B_t[t+1] - B_t[t]) / B_t[t]

        W_lev = W_lev * (1 + pi_lev*stock_ret + (1-pi_lev)*bond_ret - cost_lev)
        W_lev = torch.clamp(W_lev, min=1e-6)
        pi_prev_lev = pi_lev

    # Using Log Utility (gamma=1) to let it be hyper-aggressive
    loss_lev = -torch.mean(torch.log(W_lev))
    loss_lev.backward()
    opt_lev.step()
print("Training Complete!")
```

```python
# ==============================================================================
# BONUS BLOCK 2: REAL-WORLD BACKTEST OF THE LEVERAGED MODEL
# ==============================================================================
target_weights_lev = []
model_lev.eval()

with torch.no_grad():
    for i in range(len(df_bt)):
        target_weights_lev.append(model_lev(
            torch.tensor([0.0], dtype=torch.float32),
            torch.tensor([variance_norm.iloc[i]], dtype=torch.float32)
        ).item())

# Run through the exact same realistic trading engine
ret_lev, act_w_lev, trades_lev = run_realistic_trading(target_weights_lev, df_bt['spy_ret'], df_bt['tlt_ret'])
```

```python
# ==============================================================================
# BONUS BLOCK 3: LEVERAGED VS CONSTRAINED COMPARISON
# ==============================================================================
ann_ret_l, vol_l, sharpe_l, dd_l, cum_l, draw_l = calculate_precise_metrics(ret_lev)

print("\n" + "="*80)
print("THE HEDGE FUND TEST: UNCONSTRAINED VS CONSTRAINED")
print("="*80)
print(f"{'Metric':<25} | {'Leveraged ANN (Max 3x)':<25} | {'Constrained ANN (Max 1x)':<25}")
print("-" * 80)
print(f"{'Annualized Return':<25} | {ann_ret_l:>20.6f} | {ann_ret_r:>20.6f}")
print(f"{'Max Drawdown':<25} | {dd_l:>20.6f} | {dd_r:>20.6f}")
print(f"{'Sharpe Ratio':<25} | {sharpe_l:>20.6f} | {sharpe_r:>20.6f}")
print("="*80)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), gridspec_kw={'height_ratios': [2, 1]}, sharex=True)

# Top Panel: The massive wealth disparity
ax1.plot(df_bt.index, cum_l, label=f"Leveraged ANN (Max DD: {dd_l:.1%})", color='red', linewidth=2)
ax1.plot(df_bt.index, cum_r, label=f"Constrained ANN (Max DD: {dd_r:.1%})", color='blue', alpha=0.7, linewidth=2)
ax1.plot(df_bt.index, cum_q, label=f"SPY Benchmark", color='green', alpha=0.4)
ax1.set_yscale('log')
ax1.set_title("Wealth Accumulation: The Power (and Danger) of Unconstrained Leverage", fontsize=14)
ax1.set_ylabel("Wealth (Log Scale)")
ax1.legend(loc='upper left')
ax1.grid(alpha=0.3)

# Bottom Panel: Seeing it break the 1.0 ceiling
ax2.plot(df_bt.index, target_weights_lev, label="Leveraged Target ($\pi$)", color='red')
ax2.plot(df_bt.index, target_weights_regime, label="Constrained Target ($\pi$)", color='blue', alpha=0.6)
ax2.axhline(1.0, color='black', linestyle='--', label='100% Equity Limit (No Margin)')
ax2.set_ylabel("Target Equity Weight ($\pi$)")
ax2.legend(loc='upper left')
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ==============================================================================
# BONUS BLOCK 4: THE ISOLATED "FIREWORKS" SANDBOX
# (Does NOT modify your main project variables)
# ==============================================================================
print("Generating isolated 'cheated' market data for the AI...")

# 1. Generate new paths with massive drift (25%) and 0% risk-free rate
# We use fewer paths (1000) so it trains in seconds, not minutes.
S_cheat, v_cheat, B_cheat = simulate_heston(mu=0.25, r=0.0, paths=1000)

# Convert to distinct tensors so we don't ruin your main project's memory
S_c_t = torch.tensor(S_cheat, dtype=torch.float32)
v_c_t = torch.tensor(v_cheat, dtype=torch.float32)
B_c_t = torch.tensor(B_cheat, dtype=torch.float32)

v_c_mean = v_c_t.mean().item()
v_c_std = v_c_t.std().item() + 1e-8

# 2. Train the Leveraged Model on this cheated data
model_fireworks = PolicyANN_Leveraged() # Assuming you ran Bonus Block 1
opt_fireworks = optim.Adam(model_fireworks.parameters(), lr=0.005)

print("Training Unconstrained Model on 25% Bull Market...")
for epoch in range(100):
    opt_fireworks.zero_grad()
    W_f = torch.ones(S_cheat.shape[1])
    pi_prev_f = torch.full((S_cheat.shape[1],), 0.5)

    for t in range(252):
        t_input = torch.full((S_cheat.shape[1],), t*dt)
        var_input = (v_c_t[t] - v_c_mean) / v_c_std

        pi_f = model_fireworks(t_input, var_input)
        cost_f = torch.abs(pi_f - pi_prev_f) * fee_rate

        stock_ret = (S_c_t[t+1] - S_c_t[t]) / S_c_t[t]
        bond_ret = (B_c_t[t+1] - B_c_t[t]) / B_c_t[t]

        W_f = W_f * (1 + pi_f*stock_ret + (1-pi_f)*bond_ret - cost_f)
        W_f = torch.clamp(W_f, min=1e-6)
        pi_prev_f = pi_f

    loss_f = -torch.mean(torch.log(W_f))
    loss_f.backward()
    opt_fireworks.step()

print("Training Complete! Plotting the AI's Leveraged Strategy...")

# 3. Plot what the AI decided to do against the real VIX
model_fireworks.eval()
firework_targets = []

with torch.no_grad():
    for i in range(len(df_bt)):
        firework_targets.append(model_fireworks(
            torch.tensor([0.0], dtype=torch.float32),
            torch.tensor([variance_norm.iloc[i]], dtype=torch.float32)
        ).item())

plt.figure(figsize=(14, 5))
plt.plot(df_bt.index, firework_targets, color='red', linewidth=2, label='Unconstrained AI Target ($\pi$)')
plt.axhline(1.0, color='black', linestyle='--', label='100% Equity (Normal Limit)')
plt.axhline(3.0, color='blue', linestyle=':', label='300% Maximum Margin Cap')
plt.title("The 'Fireworks' Test: AI Given 25% Market Drift and 3x Leverage", fontsize=14)
plt.ylabel("Target Equity Weight")
plt.ylim(0, 3.2)
plt.legend(loc='lower left')
plt.grid(alpha=0.3)

# Overlay VIX to see how it scales down during crashes
ax2 = plt.gca().twinx()
ax2.plot(df_bt.index, df_bt['vix'], color='purple', alpha=0.2, label='VIX Index')
ax2.set_ylabel("VIX", color='purple')

plt.tight_layout()
plt.show()
```

```python
# ==============================================================================
# BONUS BLOCK 5: PLOTTING THE FIREWORKS WEALTH
# ==============================================================================

# 1. Run the leveraged targets through the realistic trading engine
ret_fireworks, act_w_fireworks, trades_fireworks = run_realistic_trading(
    firework_targets, df_bt['spy_ret'], df_bt['tlt_ret']
)

# 2. Calculate the precise metrics
ann_ret_f, vol_f, sharpe_f, dd_f, cum_f, draw_f = calculate_precise_metrics(ret_fireworks)

# 3. Print the metrics comparison
print("\n" + "="*80)
print("THE FIREWORKS RESULTS (3x LEVERAGE, TRAINED ON 25% DRIFT)")
print("="*80)
print(f"{'Metric':<25} | {'Fireworks Leveraged ANN':<25} | {'SPY Benchmark':<25}")
print("-" * 80)
print(f"{'Trades Executed':<25} | {trades_fireworks:<25} | {'0 (Frictionless)':<25}")
print(f"{'Annualized Return':<25} | {ann_ret_f:>25.6f} | {ann_ret_q:>25.6f}")
print(f"{'Max Drawdown':<25} | {dd_f:>25.6f} | {dd_q:>25.6f}")
print(f"{'Sharpe Ratio':<25} | {sharpe_f:>25.6f} | {sharpe_q:>25.6f}")
print("="*80)

# 4. Plot the Wealth Accumulation
plt.figure(figsize=(14, 6))

# We use a log scale because 3x leverage compounds exponentially
plt.plot(df_bt.index, cum_f, label=f"Fireworks 3x Leveraged ANN (Max DD: {dd_f:.1%})", color='red', linewidth=2)
plt.plot(df_bt.index, cum_q, label=f"SPY Benchmark (Max DD: {dd_q:.1%})", color='green', alpha=0.6, linewidth=1.5)

plt.yscale('log')
plt.title("Wealth Accumulation: Unconstrained 3x Leveraged AI vs Benchmark", fontsize=14)
plt.ylabel("Cumulative Wealth (Log Scale)")
plt.grid(alpha=0.3)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
